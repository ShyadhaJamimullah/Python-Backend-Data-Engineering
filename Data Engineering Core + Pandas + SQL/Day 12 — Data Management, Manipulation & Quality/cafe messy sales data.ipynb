```python
import pandas as pd
```

```python
df=pd.read_csv("archive (5)/dirty_cafe_sales.csv")
print(df.shape)
```

```python
df.head()
df.info()
df.isnull().sum()
```

```python
#Renaming column names 
df.columns=(df.columns.str.strip().str.lower().str.replace(" ","_"))
df.head()
```

```python
df.replace(["ERROR","UNKNOWN","N/A","","NULL","None"],pd.NA)
```

```python
for col in df.columns:
    if df[col].dtype=="object":
        df[col]=df[col].str.strip()
```

```python
for col in ["item", "payment_method", "location"]:
    df[col]=df[col].str.strip().str.title()
```

```python
#type conversion
numeric_cols=["quantity","price_per_unit","total_spent"]
for col in numeric_cols:
    df[col]=pd.to_numeric(df[col],errors="coerce")

df["transaction_date"]=pd.to_datetime(df["transaction_date"],errors="coerce")

df=df.dropna(subset=["transaction_date"])
```

```python
#handling missing values
for col in numeric_cols:
    median=df[col].median()
    df[col]=df[col].fillna(df[col].median())

cat_cols = ["item","payment_method","location"]

for col in cat_cols:
    df[col]=df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.duplicated().any()
```

```python
df_sorted=df.sort_values(by=["transaction_date","total_spent"],ascending=[True,False])
print(df_sorted)
```

```python
filtered_df=df[df["payment_method"]=="Cash"]
print(filtered_df)
```

```python
high_sales = df[df["total_spent"] > 20]
print(high_sales)
```

```python
most_bought = df.groupby("item")["quantity"].sum().sort_values(ascending=False)
print(most_bought)
```
